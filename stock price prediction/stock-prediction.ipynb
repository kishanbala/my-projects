# LSTM

```python
#Import dependencies
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import Dense 
```

```python
X = data.Open.values
Y = data.Close.values
X = np.log(X)
Y = np.log(Y)
X = X.reshape(-1,1,1)
Y = Y.reshape(-1,1,1)
x_train = X[:200000]
y_train = Y[:200000]
x_test = X[200000:]
y_test = Y[200000:]
```

```python
lstm = Sequential()

lstm.add(LSTM(units=100,return_sequences=True,input_shape=(x_train.shape[1], 1)))
lstm.add(Dropout(0.3))

lstm.add(LSTM(units=100,return_sequences=True))
lstm.add(Dropout(0.3))

lstm.add(LSTM(units=100,return_sequences=True))
lstm.add(Dropout(0.3))

lstm.add(LSTM(units=100))
lstm.add(Dropout(0.2))

lstm.add(Dense(units=1))

lstm.compile(optimizer='SGD',loss='mean_squared_error')

lst_model = lstm.fit(x_train,y_train,epochs=500,batch_size=1000,validation_split=0.3)
```

```python
plt.plot(lst_model.history['loss'])
plt.plot(lst_model.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
