```python
import pandas as pd
gender_submission = pd.read_csv("../input/titanic/gender_submission.csv")
test = pd.read_csv("../input/titanic/test.csv")
train = pd.read_csv("../input/titanic/train.csv")
```

```python
C = train.loc[train.Embarked == 'C']['Survived']
rate_embarked_C = sum(C) / len(C)

print("% of people boarded at Cherbourg survived is", rate_embarked_C)

S = train.loc[train.Embarked == 'S']['Survived']
rate_embarked_S = sum(S) / len(S)

print("% of people boarded at Southampton survived is", rate_embarked_S)

Q = train.loc[train.Embarked == 'Q']['Survived']
rate_embarked_Q = sum(Q) / len(Q)

print("% of people boarded at Queenstown survived is", rate_embarked_Q)
```

```python
train['Age'] = train['Age'].fillna(train['Age'].mean()) #Imputing missing values in Age with mean 
test['Age'] = test['Age'].fillna(test['Age'].mean()) #Imputing missing values in Age with mean 
```

```python
features = ['Pclass','Sex','Embarked']
x_train = pd.get_dummies(train[features])
y = train['Survived']
x_test = pd.get_dummies(test[features])
```

```python
#Fill NAs
new_train = x_train.fillna(method='ffill')
new_test = x_test.fillna(method='ffill')
```

```python
from sklearn.tree import DecisionTreeClassifier as DTree

classifier = DTree(random_state = 0)
classifier.fit(x_train, y)
predictions = classifier.predict(x_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
clf = AdaBoostClassifier(random_state = 1)
clf.fit(x_train, y)
pred = clf.predict(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y)
naive_pred = model.predict(x_test)
```

```python
output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': predictions})
output.to_csv('DTreewithoutAge.csv', index=False)
print("Your submission was successfully saved!")
```
